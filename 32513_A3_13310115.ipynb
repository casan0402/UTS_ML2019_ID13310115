#Project: Finding the whales in the mobile game

##Aims
In 2019, mobile will again be the largest segment, generating revenues of $68.5 billion and comprising 45% of the global games market. In 2022, the estimated proportion of mobile game’s revenue will reach 49% of global game market (Newzoo, 2019). It can be concluded that mobile game is the most important role compared to other platforms such as PC games and console games. Besides, Adweek stated that the top ten percent of spenders on a mobile game typically make up seventy percent of its in-app purchase revenue and fifty-nine percent of its total revenue. These types of players are called “Whales” in the gaming industry. As such, whales are not simply an anomaly to be ignored. They can be the driving force that helps keep a game alive and free for everyone to play (Batt, 2018). 

The aim of this project is to solve the problem that how we find the critical features of players who tend to spend large amount of money in the mobile game monthly and continuously so that the company of mobile game can make a marketing plan to increase the revenue in in-App purchase. This aim can be broken into three objectives.

1.Collect the data which generated by the players who tend to spend more money than others in a structed form.

2.Using the dataset to train the analytic model.

3.Evaluate various classifiers and choose the one that has the highest accuracy rate to predict player’s behavior.

4.Deploy the techniques to generate a report to support the marketing decision by the marketing department.

As above, it is interesting to know the behaviors that whales like to spend money in the mobile game. It is also important to increase the revenue of the mobile game company through the analysis.
