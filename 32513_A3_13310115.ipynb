#Project Title: Finding the whales in the mobile game

#1.Aims
In 2019, mobile will again be the largest segment, generating revenues of $68.5 billion and comprising 45% of the global games market. In 2022, the estimated proportion of mobile game’s revenue will reach 49% of global game market (Newzoo, 2019). It can be concluded that mobile game is the most important role compared to other platforms such as PC games and console games. Besides, Adweek stated that the top ten percent of spenders on a mobile game typically make up seventy percent of its in-app purchase revenue and fifty-nine percent of its total revenue. These types of players are called “Whales” in the gaming industry. As such, whales are not simply an anomaly to be ignored. They can be the driving force that helps keep a game alive and free for everyone to play (Batt, 2018). 

The aim of this project is to solve the problem that how we find the critical features of players who tend to spend large amount of money in the mobile game monthly and continuously so that the company of mobile game can make a marketing plan to increase the revenue in in-App purchase. This aim can be broken into three objectives.

1.Collect the data which generated by the players who tend to spend more money than others in a structed form.

2.Using the dataset to train the analytic model.

3.Evaluate various classifiers and choose the one that has the highest accuracy rate to predict player’s behavior.

4.Deploy the techniques to generate a report to support the marketing decision by the marketing department.

As above, it is interesting to know the behaviors that whales like to spend money in the mobile game. It is also important to increase the revenue of the mobile game company through the analysis.

#2.Background
The importance to find the preference of players that contribute most to the mobile game is well-known by the game producer. However, most of these producers does not understand the value of the data that they possess. The existing methods to solve this problem have two difficulties. For one thing, the data is not stored in a structured format so that this data is not viable for further analysis. For another thing, though some company stores their data in a proper way, they do not know how to use it. Sometimes these company simply transform these data to a statistical report. Then, the marketing department review the statistical report and make a marketing decision by its intuition. 

The difference between the approach of proposed project and the existing approach is that we use the technique of machine learning. As we can find the report across the publications, they are based on the statistical method. This kind of method only gives shallow information. For example, they might state that 76% of players spend money on public holiday, or 30% of players buy the diamonds and use them to accelerate building their bases. However, the department of marketing is unable to make a clear marketing plan through these kinds of reports. We can create a data analytical report which give a quantitative result to assist the marketing department to develop a clearer strategy. For example, in mid-autumn festival, send the special greetings and offers to the players who are located in China because they have 90% of possibility to spend more money during the period.

##3.3The procedure to solve domain problem
In order to solve the problem, there are four steps we will implement in the project. They are shown as below.

First, our team will collect the raw data generate from the players who are willing to spend money in the mobile game. For, example, the data includes age, gender, region, duration of playing, frequency of playing, average amount of spending per month, maximum spending, items for spending, the event of spending, etc. Then, these data will be well-organized in a structured data format and stored in a database management System such as PostgreSQL.
<br>
Second, this dataset will be used to train the model to predict the potential players who are likely to spend large amount of money in the game. The programming language we will use is Python because it has a well-developed library such Numpy, Matplotlib, Scikit-learn, and Pandas that are suitable for creating the model. Besides, we will train different classifiers including SVM, random forest, logistic regression, and naive bayes for the project.
<br>
Third, we will evaluate the accuracy of prediction and suitability of these classifiers and use it for deployment.
<br>
Finally, we will deploy this approach to the newly generated data that never used for training and create the final report as a means of supporting marketing decision for marketing department.
